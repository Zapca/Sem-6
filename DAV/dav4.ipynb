```python
import pandas as pd 
import numpy as n 
import matplotlib.pyplot as mtp
```

```python
data = pd.read_csv('DailyDelhiClimateTrain.csv')
data_dummy = data
dum2 = dum3 = pd.read_csv('DailyDelhiClimateTrain.csv')
data['humidity'].fillna(data['humidity'].mean(), inplace=True)
data['meantemp'].fillna(data['meantemp'].mean(), inplace=True)
data['meanpressure'].fillna(data['meanpressure'].mean(), inplace=True)
data['wind_speed'].fillna(data['wind_speed'].mean(), inplace=True)
```

```python
dum2['date'] = pd.to_datetime(dum2['date'])
dum2.set_index(dum2['date'].dt.to_period('M'), inplace=True)
print(dum2.head(5))
print("------------------------------------------------------------------------------------")
print(f'THe data type of the index is :- {type(dum2.index)}')
print("------------------------------------------------------------------------------------")
print(f'the values of the indes{dum2.index.values}')
```

```python
# Assuming '2013-01-01' is your reference date
start_date = pd.to_datetime('2013-01-01')
dum3['date'] = pd.to_datetime(dum3['date']) - start_date
dum3.set_index('date', inplace=True)
print(dum3.head(5))
print("------------------------------------------------------------------------------------")
print(f'THe data type of the index is :- {type(dum3.index)}')
print("------------------------------------------------------------------------------------")
print(f'the values of the indexs{dum3.index.values}')
```

```python
monthly_data = data.resample('M').mean()
monthly_data.head(5)
```

```python
yearly_data = data.resample('Y').mean()
yearly_data.head(5)
```

```python
summed_monthly_data = data.resample('M').mean()
summed_monthly_data.head(5)
```

```python
data['rolling_mean'] = data['humidity'].rolling(window=10).mean()
```

```python
data['rolling_mean'].isnull().sum()
data['rolling_mean'].fillna(data['rolling_mean'].mean())
```

```python
data.tail(5)
```

```python
dum2['humidity'] = dum2['humidity'].rolling(window=25).mean()
dum2['meantemp'] = dum2['meantemp'].rolling(window=25).mean()
dum2['wind_speed'] = dum2['wind_speed'].rolling(window=25).mean()
dum2['meanpressure'] = dum2['meanpressure'].rolling(window=25).mean()

dum2.tail(5)
```

```python
mtp.plot(dum2.index, dum2['meantemp'], label='Mean TEmp')
mtp.title('Time-series Data')
mtp.xlabel('Date')
mtp.ylabel('TEmperature')
```

```python
mtp.plot(summed_monthly_data.index, summed_monthly_data['meanpressure'], label='Mean TEmp')
mtp.title('Time-series Data')
mtp.xlabel('Date')
mtp.ylabel('Pressure')
```

```python
mtp.plot(data.index, data['rolling_mean'], label='Rolling window of Humidity')
mtp.title('Time-series Data')
mtp.xlabel('Date')
mtp.ylabel('Humidity')
```
